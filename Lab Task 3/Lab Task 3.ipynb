```python


JUG_X = 4
JUG_Y = 3
TARGET = 2

visited_states = set()

def dfs_solver(x, y, steps):
    if x == TARGET:
        print("\nGoal Achieved Successfully!\n")
        for i, step in enumerate(steps, 1):
            print(f"Step {i}: {step}")
        print(f"\nFinal State: ({x}, {y})")
        return True

    if (x, y) in visited_states:
        return False

    visited_states.add((x, y))

    possible_moves = []

    # Rule 1: Fill Jug X
    possible_moves.append((JUG_X, y, f"Fill Jug X → ({JUG_X}, {y})"))

    # Rule 2: Fill Jug Y
    possible_moves.append((x, JUG_Y, f"Fill Jug Y → ({x}, {JUG_Y})"))

    # Rule 3: Empty Jug X
    possible_moves.append((0, y, f"Empty Jug X → (0, {y})"))

    # Rule 4: Empty Jug Y
    possible_moves.append((x, 0, f"Empty Jug Y → ({x}, 0)"))

    # Rule 5: Pour X → Y
    pour_xy = min(x, JUG_Y - y)
    possible_moves.append((x - pour_xy, y + pour_xy,
                           f"Pour X → Y → ({x - pour_xy}, {y + pour_xy})"))

    # Rule 6: Pour Y → X
    pour_yx = min(y, JUG_X - x)
    possible_moves.append((x + pour_yx, y - pour_yx,
                           f"Pour Y → X → ({x + pour_yx}, {y - pour_yx})"))

    for nx, ny, action in possible_moves:
        if dfs_solver(nx, ny, steps + [action]):
            return True

    return False


dfs_solver(0, 0, [])
```
