```python


def is_safe(board, row, col, n):
    for i in range(row):
        if board[i] == col:
            return False

    # Check left diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i] == j:
            return False
        i -= 1
        j -= 1

    # Check right diagonal
    i, j = row, col
    while i >= 0 and j < n:
        if board[i] == j:
            return False
        i -= 1
        j += 1

    return True


def solve_nqueens(board, row, n):
    if row == n:
        print_solution(board, n)
        return True

    solution_found = False

    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            solution_found = solve_nqueens(board, row + 1, n) or solution_found
            board[row] = -1

    return solution_found


def print_solution(board, n):
    print("\nOne Valid Solution:\n")
    for i in range(n):
        for j in range(n):
            if board[i] == j:
                print("Q", end=" ")
            else:
                print(".", end=" ")
        print()


N = int(input("Enter number of queens (N): "))
board = [-1] * N

if not solve_nqueens(board, 0, N):
    print("No solution exists.")
```
