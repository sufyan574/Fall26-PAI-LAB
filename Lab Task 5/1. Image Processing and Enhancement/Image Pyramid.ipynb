```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('pyramids.jpg')  

if image is None:
    print("Error: Image not found! Check file name or path.")
    exit()

# Create Gaussian Pyramid
pyramid = [image]
for i in range(3):
    image = cv2.pyrDown(image)
    pyramid.append(image)

# Display Pyramid levels using matplotlib
for i in range(len(pyramid)-1, -1, -1):
    plt.figure()
    plt.title(f"Pyramid Level {i}")
    img_rgb = cv2.cvtColor(pyramid[i], cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```
