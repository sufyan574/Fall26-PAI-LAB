```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from textblob import TextBlob

print("=== Setting up NLTK Resources (run once) ===")

resources = ['punkt', 'punkt_tab', 'stopwords', 'wordnet']

for resource in resources:
    try:
        nltk.data.find(resource)
        print(f"✓ {resource} is already downloaded.")
    except LookupError:
        print(f"↓ Downloading {resource} ...")
        nltk.download(resource, quiet=False, halt_on_error=False)

print("NLTK setup completed.\n")

texts = [
    "This product is excellent and works amazingly well!",
    "I hate this movie, it was a complete waste of time.",
    "The weather today is okay, neither good nor bad.",
    "Running and playing football are my favorite activities.",
    "Better late than never, but this service is terrible."
]

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    
    stemmer = PorterStemmer()
    stemmed = [stemmer.stem(word) for word in filtered_tokens]
    
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    
    return {
        'original': text,
        'tokens': tokens,
        'filtered_tokens': filtered_tokens,
        'stemmed': stemmed,
        'lemmatized': lemmatized
    }

def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    
    if polarity > 0.05:
        sentiment = "Positive"
    elif polarity < -0.05:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    return {
        'polarity': round(polarity, 3),
        'subjectivity': round(subjectivity, 3),
        'sentiment': sentiment
    }

print("=== NLP Preprocessing and Sentiment Analysis ===\n")

for i, text in enumerate(texts, 1):
    print(f"Text {i}: {text}")
    
    processed = preprocess_text(text)
    print("  Tokens          :", processed['tokens'])
    print("  After Stopwords :", processed['filtered_tokens'])
    print("  Stemmed         :", processed['stemmed'])
    print("  Lemmatized      :", processed['lemmatized'])
    
    sentiment = analyze_sentiment(text)
    print("  Sentiment       :", sentiment['sentiment'])
    print("  Polarity        :", sentiment['polarity'])
    print("  Subjectivity    :", sentiment['subjectivity'])
    print("-" * 65)

print("\n=== Bonus: Simple Text Classification (Positive/Negative) ===")
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

train_texts = [
    "This product is excellent and works amazingly well",
    "I love this phone",
    "I hate this movie it was terrible",
    "Worst service ever",
    "Amazing experience highly recommended"
]
train_labels = ["positive", "positive", "negative", "negative", "positive"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(train_texts)
model = MultinomialNB()
model.fit(X, train_labels)

test_text = "This service is terrible and slow"
prediction = model.predict(vectorizer.transform([test_text]))[0]
print(f"Text: '{test_text}' → Predicted: {prediction}")
```
